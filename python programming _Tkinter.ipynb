```python
from tkinter import*
from tkinter import Button
win =Tk()
win.geometry('700x300')
win.title('MY APP')

def np():
    file=Toplevel(win)
    file.geometry('300x200')
    b=Button(file ,text='b1',padx=20,pady=10,activebackground='black',activeforeground='white',bg='yellow')
    b.place(x=20,y=20)
    file.title('MY PAGE')
mb= Menubutton(win,text='file')
mb.grid()
mb.menu=Menu(mb)
mb['menu']=mb.menu
mb.menu.add_command(label='window',command=np)
mb.menu.add_command(label='file',command=np)
mb.menu.add_separator()
mb.menu.add_command(label='open',command=np)
mb.menu.add_command(label='close',command=np)
mb.menu.add_separator()
mb.menu.add_command(label='save',command=np)
mb.menu.add_command(label='go to',command=np)
mb.menu.add_command(label='close',command=win.destroy)

mb.configure(bg='blue')

win.mainloop()
```

```python
from tkinter import *
window = Tk()

def convert():
    value = float(ft_entry.get())
    meter = value * 0.3048
    mt_value.set("%.4f"% meter)


def clear():
    ft_value.set(" ")
    mt_value.set(" ")


window.title("F2M APP")
window.config(background ="light green")
window.geometry("320x220")
window.resizable(width=False,height=False)

ft_label = Label(window,text="Feet",bg="purple",fg="white",width=14)
ft_label.grid(column=0,row =0, padx=15, pady=15)

ft_value = StringVar()
ft_entry = Entry(window,textvariable=ft_value,width= 14)
ft_entry.grid(column=1,row=0)


mt_label = Label(window,text="Meter",bg="brown",fg="white",width=14)
mt_label.grid(column=0,row =1, padx=15, pady=15)

mt_value = DoubleVar()
mt_entry = Entry(window,textvariable=mt_value,width= 14)
mt_entry.grid(column=1,row=1)
mt_entry.delete(0,'end')

convert_btn = Button(window,text="Convert",bg="blue",fg="white",width=14,command=convert)
convert_btn.grid(column=0,row=3,padx=15)

convert_btn = Button(window,text="Clear",bg="black",fg="white",width=14,command=clear)
convert_btn.grid(column=1,row=3,padx=15)

window.mainloop()
```

```python
from tkinter import *
class TipCalc:
    def __init__(self):
        win =Tk()
        win.geometry('500x400')
        win.title('covid-19 vaccine registration')
        win.config(bg='sky blue')
        win.resizable(width=False,height=False)
        self.tipp=IntVar()
        self.amt=StringVar()
        self.dis=StringVar()
        self.bill=StringVar()
        Label(win,text='Tip Percentage',bg='purple',fg='white',width=15).grid(column=0,row=0,padx=15, pady=15)
        Label(win,text='Total Amount',bg='purple',fg='white',width=15).grid(column=1,row=1,padx=15, pady=15)
        Label(win,text='Discount Amount',bg='purple',fg='white',width=15).grid(column=1,row=2,padx=15, pady=15)
        Label(win,text='Total Bill',bg='purple',fg='white',width=15).grid(column=1,row=3,padx=15, pady=15)
        e1=Entry(win,width=30,textvariable=self.amt)
        e1.grid(column=2,row=1)
        e2=Label(win,width=30,bg='sky blue',textvariable=self.dis)
        e2.grid(column=2,row=2)
        e3=Label(win,width=30,bg='sky blue',textvariable=self.bill)
        e3.grid(column=2,row=3)
        
        rb1=Radiobutton(win,text='5%',value=5,variable=self.tipp)
        rb1.grid(column=0,row=1,padx=15, pady=15)
        rb2=Radiobutton(win,text='10%',value=10,variable=self.tipp)
        rb2.grid(column=0,row=2,padx=15, pady=15)
        rb3=Radiobutton(win,text='15%',value=15,variable=self.tipp)
        rb3.grid(column=0,row=3,padx=15, pady=15)
        rb4=Radiobutton(win,text='20%',value=20,variable=self.tipp)
        rb4.grid(column=0,row=4,padx=15, pady=15)
        rb5=Radiobutton(win,text='25%',value=25,variable=self.tipp)
        rb5.grid(column=0,row=5,padx=15, pady=15)
        
        
            
        b1=Button(win,text='Calculate',bg='black',fg='white',width=15,command=self.calculate)
        b1.grid(column=1,row=4)
        b2=Button(win,text='Clear',bg='black',fg='white',width=15,command=self.clear)
        b2.grid(column=2,row=4)


        win.mainloop()
        
    def calculate(self):
        pre_tip = float(self.amt.get())
        percentage =int( self.tipp.get() )/ 100
        tipa= pre_tip * percentage
        self.dis.set(tipa)

        final_bill = pre_tip + tipa
        self.bill.set(final_bill)

    def clear(self):
        self.amt.set("")
        self.bill.set("")
        self.dis.set("") 

        
TipCalc()
```
